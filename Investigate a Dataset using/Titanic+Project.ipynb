there are some changes after removing passenngers with missing ages. For example,the surviving rate incresesd by 2.23%. This  means that most of missing ages for unsurvived passengers it maybe because we can gather data from survived passenger. 

```python
#Updated_Passengers = Passengers['Age'].fillna(29.7)
print ('Before' , len(Passengers))
Updated_Passengers = Passengers.dropna(subset = ['Age'])
print ('After' , len(Updated_Passengers))
```
